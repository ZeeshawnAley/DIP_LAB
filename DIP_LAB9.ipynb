# Task 1

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('img.png')

horizontal_mask = np.array([[-1,-1,-1], [2,2,2], [-1,-1,-1]])
vertical_mask = np.array([[-1,2,-1], [-1,2,-1], [-1,2,-1]])
diagonal_mask1 = np.array([[2,-1,-1], [-1,2,-1], [-1,-1,2]])
diagonal_mask2 = np.array([[-1,-1,2], [-1,2,-1], [2,-1,-1]])

horizontal = cv2.filter2D(image, -1, horizontal_mask)
vertical = cv2.filter2D(image, -1,vertical_mask)
diagonal1 = cv2.filter2D(image, -1,diagonal_mask1)
diagonal2 = cv2.filter2D(image, -1,diagonal_mask2)

plt.subplot(2, 2, 1)
plt.imshow(horizontal)
plt.title('Horizontal Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(vertical)
plt.title('Vertical Image')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(diagonal1)
plt.title('-45 Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(diagonal2)
plt.title('+45 Image')
plt.axis('off')
```

# Task 2 (In-built)

```python
image1 = cv2.imread('img1.jpg')
image1 = cv2.GaussianBlur(image1, (3,3), 0)

sobel_x = cv2.Sobel(image1, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image1, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = cv2.Sobel(image1, cv2.CV_64F, 1, 1, ksize=3)

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image1,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Horizontal Edge')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Vertical Edge')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Combined')
plt.axis('off')

plt.show()
```

# Task 2 (Filter)

```python
image1 = cv2.imread('img1.jpg')
image1 = cv2.GaussianBlur(image1, (3,3), 0)

horizontal_mask = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])
vertical_mask = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
diagonal_mask1 = np.array([[2,-1,-1], [-1,2,-1], [-1,-1,2]])

sobel_x = cv2.filter2D(image1, -1, horizontal_mask)
sobel_y = cv2.filter2D(image1, -1, vertical_mask)
magnitude = cv2.filter2D(image1, -1, diagonal_mask1)

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image1,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Horizontal Edge')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Vertical Edge')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Combined')
plt.axis('off')

plt.show()
```

# Task 3 (In-built)

```python
image2 = cv2.imread('img2.png', 0)
image2 = cv2.GaussianBlur(image2, (3,3), 0)

canny = cv2.Canny(image2, threshold1 = 50, threshold2 = 150)

plt.subplot(1, 2, 1)
plt.imshow(image2, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(canny, cmap='gray')
plt.title('Output Image')
plt.axis('off')
```
